```python
# 1 load 2 info 3 resize 4 check
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('/Users/xh/workspaces/jupyter/opencv-example/2入门/opencv入门/image0.JPG', 1)
imgInfo = img.shape
print(imgInfo)
height = imgInfo[0]
width = imgInfo[1]
mode = imgInfo[2]
# 1 放大 缩小 2 等比例 非 2:3 
dstHeight = int(height*0.5)
dstWidth = int(width*0.5)
#最近临域插值 双线性插值 像素关系重采样 立方插值
dst = cv2.resize(img,(dstWidth,dstHeight))
print(dst.shape)
img_rgb = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
import matplotlib.pyplot as plt
img_rgb = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('Image')
plt.axis('off')
plt.show()
plt.title('Image')
plt.axis('off')
plt.show()
```
