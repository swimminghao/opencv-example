```python
import tensorflow as tf
data1 = tf.constant(6)
data2 = tf.constant(2)
dataAdd = tf.add(data1,data2)
dataMul = tf.multiply(data1,data2)
dataSub = tf.subtract(data1,data2)
dataDiv = tf.divide(data1,data2)

print(dataAdd)
print(dataMul)
print(dataSub)
print(dataDiv)
print('end!')
```

```python
import tensorflow as tf
data1 = tf.constant(6)
data2 = tf.Variable(2)
dataAdd = tf.add(data1,data2)
dataCopy = tf.compat.v1.assign(data2,dataAdd)# dataAdd ->data2
dataMul = tf.multiply(data1,data2)
dataSub = tf.subtract(data1,data2)
dataDiv = tf.divide(data1,data2)
print(dataAdd)
print(dataMul)
print(dataSub)
print(dataDiv)
print('sess.run(dataCopy)',dataCopy)#8->data2
print('dataCopy.eval()',dataCopy)#8+6->14->data = 14
print('tf.get_default_session()',dataCopy)
print('end!')
```
