```python
import cv2
import os

# 1. 读取第一张图片获取尺寸
img = cv2.imread('image1.jpg')
if img is None:
    print("❌ 错误：无法读取 image1.jpg")
    print("   请确保文件存在且路径正确")
    exit()

height, width = img.shape[:2]
size = (width, height)
print(f"视频尺寸: {width} x {height}")

# 2. 创建视频写入对象
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
videoWrite = cv2.VideoWriter('2.mp4', fourcc, 5, size)

# ✅ 检查是否创建成功
if not videoWrite.isOpened():
    print("❌ 错误：无法创建视频文件")
    print("   可能原因：编码器不支持、路径无权限")
    exit()

print("✅ 视频创建成功，开始写入...")

# 3. 写入图片
success_count = 0
for i in range(1, 11):
    fileName = f'image{i}.jpg'

    # ✅ 检查文件是否存在
    if not os.path.exists(fileName):
        print(f"⚠️ 警告：{fileName} 不存在，跳过")
        continue

    img = cv2.imread(fileName)

    # ✅ 检查图片是否读取成功
    if img is None:
        print(f"⚠️ 警告：{fileName} 读取失败，跳过")
        continue

    # ✅ 检查图片尺寸是否一致
    if img.shape[:2] != (height, width):
        print(f"⚠️ 警告：{fileName} 尺寸不一致，自动调整")
        img = cv2.resize(img, size)

    videoWrite.write(img)
    success_count += 1
    print(f"✅ 已写入: {fileName} ({success_count}/10)")

# 4. ✅ 释放资源（重要！）
videoWrite.release()
print(f"\n✅ 视频合成完成！")
print(f"   输出文件: 2.mp4")
print(f"   成功写入: {success_count} 帧")
print(f"   帧率: 5 FPS")
print(f"   时长: {success_count/5:.1f} 秒")
```

```python
import cv2
import os

# 1. 获取图像尺寸
img = cv2.imread('image1.jpg')
if img is None:
    print("错误：无法读取 image1.jpg")
    exit()

height, width = img.shape[:2]
size = (width, height)

# 2. 创建视频写入对象
# ✅ 使用 'mp4v' 编码器（通用）
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
videoWrite = cv2.VideoWriter('output.mp4', fourcc, 5, size)

# 3. 循环写入图片
for i in range(1, 11):
    fileName = f'image{i}.jpg'

    # ✅ 检查文件是否存在
    if not os.path.exists(fileName):
        print(f"警告：{fileName} 不存在，跳过")
        continue

    img = cv2.imread(fileName)
    if img is None:
        print(f"警告：{fileName} 读取失败，跳过")
        continue

    # ✅ 确保图片尺寸一致
    if img.shape[:2] != (height, width):
        img = cv2.resize(img, size)

    videoWrite.write(img)
    print(f"已写入: {fileName}")

# 4. ✅ 释放资源
videoWrite.release()
print("视频合成完成！")
```
