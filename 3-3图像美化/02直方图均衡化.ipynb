```python
#灰度 直方图均衡化 提高亮度
import cv2
import numpy as np

img = cv2.imread('/Users/xh/workspaces/jupyter/opencv-example/2入门/opencv入门/image0.JPG', 1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
dst = cv2.equalizeHist(gray)
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(15, 8))
# 原灰度图
axes[0].imshow(gray, cmap='gray')  # ✅ 指定灰度颜色映射
axes[0].set_title('Original Gray Image')
axes[0].axis('off')

# 均衡化后的图像
axes[1].imshow(dst, cmap='gray')  # ✅ 指定灰度颜色映射
axes[1].set_title('Equalized Image')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
#彩色 直方图均衡化
import cv2
import numpy as np

img = cv2.imread('image0.JPG', 1)
(b, g, r) = cv2.split(img)  #通道分解
bH = cv2.equalizeHist(b)
gH = cv2.equalizeHist(g)
rH = cv2.equalizeHist(r)
result = cv2.merge((bH, gH, rH))  # 通道合成
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(15, 8))
# 原灰度图
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # ✅ 指定灰度颜色映射
axes[0].set_title('Original Image')
axes[0].axis('off')

# 均衡化后的图像
axes[1].imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))  # ✅ 指定灰度颜色映射
axes[1].set_title('Equalized Image')
axes[1].axis('off')

plt.tight_layout()
plt.show()

```

```python
#YUV 直方图均衡化 提高亮度 不损失色彩
import cv2
import numpy as np

img = cv2.imread('./Users/xh/workspaces/jupyter/opencv-example/2入门/opencv入门/image0.JPG', 1)
imgYUV = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
channelYUV = cv2.split(imgYUV)
channelYUV[0] = cv2.equalizeHist(channelYUV[0])
channels = cv2.merge(channelYUV)
result = cv2.cvtColor(channels, cv2.COLOR_YCrCb2BGR)
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(15, 8))
# 原灰度图
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # ✅ 指定灰度颜色映射
axes[0].set_title('Original Image')
axes[0].axis('off')

# 均衡化后的图像
axes[1].imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))  # ✅ 指定灰度颜色映射
axes[1].set_title('Equalized Image')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
